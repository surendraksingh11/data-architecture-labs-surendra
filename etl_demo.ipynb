```python
from pyspark.sql.functions import split, col, when

df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("s3://els-aws-user14/")

df = df.withColumn(
        "First Name",
        when(col("Authors").isNotNull(), split(col("Authors"), " ").getItem(0))
    ).withColumn(
        "Last Name",
        when(col("Authors").isNotNull(), split(col("Authors"), " ").getItem(1))
    )

display(df)
```
